Imbalanced Dataset:<br>
A dataset with an un-equal class distribution

```python
# importing the dependencies/libraries
import pandas as pd
import numpy as np
```

```python
# loading the dataset to pandas DataFrame
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# printing the first 5 rows of DataFrame
credit_card_data.head()
```

```python
# printing the last 5 rows of DataFrame
credit_card_data.tail()
```

```python
# distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced dataset<br>
0 --> Legit Transactions<br>
1 --> Fraudulent Transactions

```python
# seperating the legit and fraudulent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-Sampling

Building a sample dataset containing similar distribution of Legit and Fraudulent transactions

Number of Fraudulent Transactions --> 177

```python
legit_sample = legit.sample(n=177)
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```
