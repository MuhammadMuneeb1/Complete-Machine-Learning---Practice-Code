Seaborn:



*   Data Visualization Library



Importing the Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Note: Seaborn has some built-in Datasets

```python
# Total Bill vs Tip Dataset
tips = sns.load_dataset('tips') # Seaborn (Tips) Dataset imported as pandas DataFrame
```

```python
tips.head()
```

```python
# Setting the Theme for the Seaborn Plots
sns.set_theme()
```

```python
# Visualize the (Tips) Data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# House price Dataset
file_path = '/content/BostonHousing.csv'
house_boston = pd.read_csv(file_path)
house = pd.DataFrame(house_boston, columns=house_boston.columns)
house.head()
```

```python
print(house_boston)
```

Distribution Plot

```python
sns.distplot(house['price'])
```

Correlation:



*   Positive Correlation: If one value increase other value also increase then they are positively correlated.

*   Negative Correlation: If one value increase and other value decrease then they are negatively correlated.



Heat Map

```python
# Constructing Heat Map to find Correlation
correlation = house.corr()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')

# More the Dark color means more the positive correlation.
# More the Light color means more the Negative correlation.
```
